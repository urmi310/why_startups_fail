```python
import pandas as pd

df = pd.read_csv('startup_failure_prediction.csv')
df.head()
```

```python

df['Startup_Status'] = df['Startup_Status'].astype(str).str.strip()
df['Startup_Status'].unique()
```

```python
df['Startup_Status'] = df['Startup_Status'].astype(int)

```

```python
df_failed = df[df['Startup_Status'] == 1]
df_failed.shape

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

industry_counts = df_failed['Industry'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=industry_counts.values, y=industry_counts.index, palette='Oranges_r')
plt.title('Top 10 Industries Where Startups Failed')
plt.xlabel('Number of Failed Startups')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()

```

```python
# Plot: Funding vs. Startup Status
plt.figure(figsize=(8,6))
sns.boxplot(x='Startup_Status', y='Funding_Amount', data=df)
plt.title('Funding Amount by Startup Status (0=Operating, 1=Failed)')
plt.xlabel('Startup Status')
plt.ylabel('Funding Amount')
plt.tight_layout()
plt.show()
```

```python
# Plot: Burn Rate vs. Startup Status
plt.figure(figsize=(8,6))
sns.boxplot(x='Startup_Status', y='Burn_Rate', data=df)
plt.title('Burn Rate by Startup Status')
plt.xlabel('Startup Status')
plt.ylabel('Burn Rate')
plt.tight_layout()
plt.show()
```

```python
# Plot: Customer Retention Rate vs. Startup Status
plt.figure(figsize=(8,6))
sns.boxplot(x='Startup_Status', y='Customer_Retention_Rate', data=df)
plt.title('Customer Retention by Startup Status')
plt.xlabel('Startup Status')
plt.ylabel('Retention Rate (%)')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# âœ… Conclusion
# - Most failed startups are in a few industries.
# - Higher burn rate & lower retention often lead to failure.
# - Funding alone doesn't guarantee success.
# - Market size and uniqueness may play important roles.
```
